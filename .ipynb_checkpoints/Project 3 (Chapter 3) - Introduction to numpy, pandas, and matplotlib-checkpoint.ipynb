```python
#zerosOnesAndLike.py  
import numpy as np  
list_of_lists = [[1,2,3], [4,5,6], [7,8,9]]  
array_of_arrays = np.array(list_of_lists)  
zeros_like_array = np.zeros_like(list_of_lists)  
ones_like_array = np.ones_like(list_of_lists)  
empty_like_array = np.empty_like(list_of_lists)  
 
print("list_of_lists:", list_of_lists, sep="\n")  
print("array_of_arrays:", array_of_arrays, sep="\n")  
print("zeros_like_array:", zeros_like_array, sep="\n")  
print("ones_like_array:", ones_like_array, sep="\n")  
print("empty_like_array:", empty_like_array, sep="\n") 
```

### Create a DataFrame from a dictionary

```python
macro_dict = {"GDP": {},
             "Money": {},
             "Real GDP": {},
             "Price Level": {}}
macro_dict
```

```python
for key in macro_dict:
    for i in range(1990, 2018):
        macro_dict[key][i] = round(np.random.random()* 10000,2)
print(macro_dict)
```

```python
pd.DataFrame(macro_dict)
```

```python
macro_df = pd.DataFrame(macro_dict)
macro_df.tail()
```

```python
macro_df["Price Level"] = macro_df["GDP"].div(macro_df["Real GDP"])
macro_df["Velocity"] = macro_df["GDP"].div(macro_df["Money"])
macro_df.head()
```

The .tail and .head is a convenient method to call the first or last portion of the DataFrame

```python
macro_df.loc[1995:2000, ["Real GDP", "Money"]]
```

```python
macro_df.loc[[1995, 1997, 1999], ["Real GDP", "Money"]]
```

```python
macro_df.loc[[year for year in range(1990, 2018, 4)], ["Real GDP", "Money"]]
```

```python
macro_df.loc[:, ["Real GDP"]]
```

```python
macro_df.loc[macro_df.index, macro_df.keys()]
```

## Plotting with Matplotlib

```python
import matplotlib.pyplot as plt
macro_df.plot()
```

```python
macro_df.plot(logy=True)
```

```python
macro_df.plot(logy=True, ls = "", marker = ".")
#preferred way of plotting
```

```python
plt.rcParams.update({"font.size" : 30})
fig, ax =  plt.subplots(figsize =(20,10))
macro_df.plot(logy=True, ls = "", marker = ".", ax = ax, markersize = 25) 
```

```python
fig, ax = plt.subplots(figsize =(20,10))
macro_df[["Real GDP", "Price Level", "Velocity"]].plot(
    #logging y allows the y axis to represent proportional change
    logy =True,
    secondary_y = "Real GDP",
    linewidth = 4, 
    # alpha is transparency 
    # between 0 and 1
    alpha = .25,
    ax = ax)
```

```python
for key, series in macro_df.items():
    print(key, series, sep ="\n")
```

```python
for key, series in macro_df.items():
    fig, ax = plt.subplots(figsize =(20,10))
    series.plot.line()
    plt.title(key)
```

```python
fig, ax = plt.subplots(figsize =(20,10))
for key, series in macro_df.items():
    series.plot.line(ax = ax, logy = True)
ax.set_title("Artificial Macro Data")
ax.set_ylabel("Dollars")
ax.set_xlabel("Year")
#ax.legend
```
