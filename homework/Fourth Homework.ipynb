## 1.

```python
import random
samples = [random.normalvariate(2,.5) for j in range(100)]
samples
```

```python
import pandas as pd
import matplotlib.pyplot as plt
samples = pd.DataFrame(samples)
samples.plot.hist(bins = 40)

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("100 Observations from a Normal Distribution")
```

```python
samples.plot(ls = "", marker = ".", markersize = 4)
```

## 2.

```python
import numpy as np
cumulative_sum = np.cumsum(samples)
cumulative_sum
```

```python
plt.plot(cumulative_sum)
plt.xlabel("Observation")
plt.ylabel("Cumulative Sum")
plt.title("Cumulative Sum of 100 Observations from a Normal Distribution")
plt.show()
```

There does appear to be a trend because the line is uppward sloping and is relatively straight. You can also tell there is a positive trend by looking at the data that gets bigger as the observations continue. 

## 3.

```python
import pandas_datareader.data as web
import yfinance as yf
import datetime

yf.pdr_override()
start = datetime.datetime(2012,1,1)
end = datetime.datetime.today()

data_dict = {}
stocks = ["AMZN", "GOOGL", "NVDA", "COST"]
for key in stocks:
    data_dict[key] = web.get_data_yahoo(key, start, end)
```

```python
data_dict["AMZN"]
```

```python
data_dict["GOOGL"]
```

```python
data_dict["NVDA"]
```

```python
data_dict["COST"]
```

```python
close_data = pd.concat([val["Close"] for val in data_dict.values()],
                      keys = data_dict.keys(),
                      axis = 1)
close_data
```

### Histograms

```python
fig, ax = plt.subplots(figsize = (15,10))
close_data["AMZN"].hist(bins = 50, label = "AMZN Data")
plt.legend()
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Amazon Stock Close Data")
```

```python
fig, ax = plt.subplots(figsize = (15,10))
close_data["GOOGL"].hist(bins = 50, label = "GOOGL Data")
plt.legend()
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Alphabet Stock Close Data")
```

```python
fig, ax = plt.subplots(figsize = (15,10))
close_data["NVDA"].hist(bins = 50, label = "NVDA Data")
plt.legend()
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("NVIDIA Corp. Stock Close Data")
```

```python
fig, ax = plt.subplots(figsize = (15,10))
close_data["COST"].hist(bins = 50, label = "COST Data")
plt.legend()
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Costco Wholesale Corp. Stock Close Data")
```

### Cumulative Sums

```python
cumulative_sum_data = np.cumsum(close_data)
cumulative_sum_data
```

```python
plt.plot(cumulative_sum_data["AMZN"])
plt.xlabel("Observation")
plt.ylabel("Cumulative Sum")
plt.title("Cumulative Sum of Amazon Data")
plt.show()
```

```python
plt.plot(cumulative_sum_data["GOOGL"])
plt.xlabel("Observation")
plt.ylabel("Cumulative Sum")
plt.title("Cumulative Sum of Alphabet Data")
plt.show()
```

```python
plt.plot(cumulative_sum_data["NVDA"])
plt.xlabel("Observation")
plt.ylabel("Cumulative Sum")
plt.title("Cumulative Sum of NVIDIA Corp. Data")
plt.show()
```

```python
plt.plot(cumulative_sum_data["COST"])
plt.xlabel("Observation")
plt.ylabel("Cumulative Sum")
plt.title("Cumulative Sum of Costco Wholesale Corp. Data")
plt.show()
```

All of these cumulative sum graphs demonstrate an upward sloping trend although they are not all at the same rate or level. Some (such as Costco) rise quicker than others (like NVIDIA Corp.) and spike in different years.
