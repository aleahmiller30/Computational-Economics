1. Generate data for one random variable with a mean of 5 and standard deviation of 1. Run a one-sample t-test that the mean of the sample is different than 0. Perform the same test again with a mean of 1 and a standard deviation of 1.

2. Why has the result changed? In answering this question, plot both samples with the index as the x-axis values. Use two colors to differentiate the samples. Indicate the value of zero using a horizontal line. Reference the plot in your answer.

3. Now run a t-test for the two samples. Print the results and explain their meaning.

## 1.

```python
import numpy as np
import random 

first_data = np.random.normal(5,1,50)
first_data
```

```python
import scipy.stats as stats
from scipy.stats import ttest_1samp

test_result_1 = ttest_1samp(first_data, 0)

print("T-Value:", test_result_1.statistic)
print("P-Value:", test_result_1.pvalue)
```

```python
second_data = np.random.normal(1, 1, 50)
second_data
```

```python
test_result_2 = ttest_1samp(second_data, 0)

print("T-Value:", test_result_2.statistic)
print("P-Value:", test_result_2.pvalue)
```

## 2.

```python
import matplotlib.pyplot as plt
import pandas as pd

fig, ax = plt.subplots(figsize = (20,10))
plt.plot(first_data, label = "Mean = 5", color = "red")
plt.plot(second_data, label = "Mean = 1", color = "blue")
plt.axhline(y=0, color='k', linestyle='--', label='Zero')
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Samples with Mean Comparison")
```

The results of the test changed because the mean changed. Using the equation,    $$t=\frac{\bar{X} - \mu}{\frac{s}{\sqrt{n}}}$$    it can be seen that the bigger the mean (working with larger numbers) the bigger the T value will be. As is shown in the graph above, the first test with a mean of five resulted in larger values whereas the results from when the mean was changed to 1 are much smaller.

## 3.

```python
stats.ttest_ind(a = first_data, b = second_data, equal_var=True)
```

This means that we do not reject the null hypothesis because the p value is greater than 0.05. The null hypothesis was that the true means do not differ.
