```python
import random
random_list1 = [random.randint(0,100) for i in range(1000)]
print(random_list1)
```

## 2.

```python
def total(lst):
    total_ = 0  
    for val in lst:
        total_ += val
    return total_
```

```python
def mean(lst):
    n= len(lst)
    mean_ = total(lst) / n
    return mean_
```

```python
def variance(lst, sample = False):
    list_mean = mean(lst)
    n = len(lst)
    DoF = n - 1
    
    sum_sq_diff = 0
    for val in lst:
        sum_sq_diff += (val - list_mean) ** 2
    if sample == False:
        variance_ = sum_sq_diff / n
    else:
        variance_ = sum_sq_diff / DoF
    return variance_
print("Variance of random_list1:", variance(random_list1))
```

## 3.

```python
nested_dict = {}
for i in range(2, 11):
    random_numbers = [random.randint(0, 100) for i in range(1000)]
    nested_dict[i] = {"list": random_numbers}
print(nested_dict)
```

## 4.

```python
import statistics
for i in range(2, 11):
    list_variance = statistics.variance(nested_dict[i]["list"])
    nested_dict[i]["variance"] = list_variance
for i in range(2, 11):
    print(f"{nested_dict[i]['variance']}")
```

## 5.

```python
def min_and_max(random_list1):
    min_value = random_list1[0]
    max_value = random_list1[0]
    
    for val in random_list1[1:]:
        if val < min_value:
            min_value = val
    
        if val > max_value:
            max_value = val
    return min_value, max_value

min_val, max_val = min_and_max(random_list1)
print(f"{min_val}")
print(f"{max_val}")
```

```python
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def random_numbers_and_plot():
# Generate random numbers
    random_numbers = [random.randint(0, 100) for _ in range(1000)]

# Matplotlib Plot
    plt.plot(random_numbers)
    plt.title('Random Numbers Plot (Matplotlib)')
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.show()

# Plotly Interactive Chart
    trace = go.Scatter(x=list(range(1000)), y=random_numbers, mode='lines', name='Random Numbers')

    layout = go.Layout(
        title='Random Numbers Plot (Plotly)',
        xaxis=dict(title='Index'),
        yaxis=dict(title='Value'),)

    fig = go.Figure(data=[trace], layout=layout)

    fig.show()
random_numbers_and_plot()
```

```python
import statistics
import plotly.graph_objects as go

def calculate_variances_and_plot(nested_dict):
    variances = []
    indexes = list(range(2, 11))

    for i in indexes:
        list_variance = statistics.variance(nested_dict[i]["list"])
        nested_dict[i]["variance"] = list_variance
        variances.append(list_variance)

    # Create an interactive bar chart using Plotly
    fig = go.Figure(data=[go.Bar(x=indexes, y=variances, text=[f"{v:.2f}" for v in variances],
                                 hoverinfo='text', marker_color='skyblue')])

    # Customize the layout for better visualization
    fig.update_layout(title="Variances of nested_dict",
                      xaxis=dict(title='Index'),
                      yaxis=dict(title='Variance'),
                      template='plotly_dark')

    # Show the interactive chart
    fig.show()
calculate_variances_and_plot(nested_dict)
```
